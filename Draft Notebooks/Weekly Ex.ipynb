```python
def shorten(text):
    """ Return a 10-character version of the string if it is longer.

    If text is longer than 10 characters, return the first four characters
    followed by '..' followed by the last four characters.
    If text is not longer than 10 characters, returns text.

    Example:
    shorten('1234567890abcd') returns '1234..abcd'
    """
    if len(text) > 10:
        return text[:4] + '..' + text[-4:]
    else:
        return text


print(shorten('1234'))
print(shorten('1234567890'))
print(shorten('1234567890abcd'))
```

```python
import os


def create_dir(dir_name):
    """Create a directory dir_name in the current working directory

    Example:
    If the current directory is c:/foo/, create_dir('bar') creates
    a c:/foo/bar directory.
    """
    os.makedirs(os.path.join(dir_name))
    


def change_dir(relative_path):
    """Change to the directory dir_name relative to the current working directory

    Example:
    If the current directory is c:/foo/, change_dir('bar') changes
    current working directory to c:/foo/bar
    """
    os.chdir(os.path.join(relative_path))


def print_current():
    """Print current working directory"""
    print(os.getcwd())


def change_to_parent_dir():
    """Change to the parent directry of the current working directory

    Example:
    If the current directory is c:/foo/bar, change_to_parent_dir() changes
    the working directory to c:/foo/
    """
    os.chdir('..')


def check(dir_name):
    create_dir(dir_name)
    print_current()
    change_dir(dir_name)
    print_current()
    change_to_parent_dir()
    print_current()


check('foo')
check('bar')
```

```python
import os

def get_file_content_as_list(filename):
    """Return content of the file as a list of lines

    The file is expected to be in the current working directory.
    The lines in the list contains trailing line breaks.

    Example:
    If the a.txt has two lines 'aaa' and 'bbb',
    get_file_content_as_list('a.txt') returns ['aaa\n', 'bbb']
    """
    items = []
    file_path = os.path.join(filename)
    f = open(file_path, 'r')
    items = f.readlines()
    f.close()
    return items
    

def get_file_stats(contents):
    """Given a list of lines, return line count and letter count as a dictionary

    Trailing line breaks (if any) are not counted for letter count.
    Spaces, even trailing spaces, are counted for letter count.
    Example:
    get_stats(['aaa\n', 'bbb']) returns {'lines': 2, 'letters': 6}
    """
    stats = []
    lineCount = 0
    letterCount = 0
    
    for line in contents:
        lineCount += 1
        for char in line:
            if char != "\n":
                letterCount += 1
            
    stats.append(lineCount)
    stats.append(letterCount)
    return stats


def analyze_file(filename):
    contents_as_list = get_file_content_as_list(filename)
    print('lines in file:', contents_as_list)
    stats = get_file_stats(contents_as_list)
    print('It has', stats[0], 'lines containing', stats[1], 'letters')


analyze_file('file1.txt')
analyze_file('file2.txt')
```

```python
def process():
    print_welcome()
    grade_report = read_entries()
    print_grades(grade_report)


def print_welcome():
    print('-----------------------------------------------------')
    print('     ~ Welcome to Grade Report System ~')
    print('* Enter one entry at a time in the format NAME GRADE')
    print('* NAME should be one word')
    print('* Enter an empty line to proceed to printing report')
    print('-----------------------------------------------------')


def read_entries():
    all_grades = {}
    while True:
        entry = input('Student name and grade?')
        if entry == "":
            break
        
        l = entry.split(' ')
        if len(l) == 2:
            name = l[0]
            grade = l[1]
            
            if grade == "":
                print("Invalid input! Entry rejected.")
                continue
            
        else:
            print("Invalid input! Entry rejected.")
            continue
            
        add_entry(all_grades, name, grade)
    return all_grades


def add_entry(grades, name, grade):
    if name in grades.keys():
        grades[name] += [grade]
    else:
        grades[name] = [grade]

def print_grades(grades):
    print('============ Grade Report ===============')
    for key in grades.keys():
        print(key, ":", " ".join(grades[key]))


process()
```

```python
def print_inventory(items, title, width):
    print_header(width, title)
    for k in sorted(items.keys()):
        print_item(k, items[k], width)
    print_total(calculate_total(items), width)

def print_header(width, title):
    """ print the given title in the format of a banner.

    Example: print_header(10, 'hi')
    ##########
    ### HI ###
    ##########

    * The title is center aligned and in upper case

    Arguments:
    word -- (type:string) the title of the banner
    width -- (type:int) the width of the banner
    """
    title = " " + title.upper() + " "
#     print("'{}'".format(title))
    filler = int((width - len(title))/2)
    print('#'*width)
    print('#'*filler + title + '#'*filler)
    print('#'*width)


def no_of_dots(name, width):
    digits = 3
    
#     while var > 0:
#         var = var//10
#         digits += 1
        
    return(width - digits - len(name) - 1)
    
    
    
def print_item(name, count, width):
    """ Print the name and count as a row in the inventory listing.

    Example: print_item('pins', 2, 12)
    pins....:  2

    * The count is right-justified in a column that is 3 characters wide.

    Arguments:
    name -- (type:string) name of the item
    count -- (type:int) number of items
    width -- (type:int) the width to use for printing
    """
#     print("name = {}, count = {}, width = {}".format(name, count, width))
    dots = 0
    dots = no_of_dots(name, width)
    print(name + "."*dots + ":" + str(count).rjust(3))
    
def print_total(total, width):
    """ print the given total in the following format

    Example: print_total(5, 12)
           ---
    total...:  5
           ===

    * The total is right-justified

    Arguments:
    total -- (type:int) the total to print
    width -- (type:int) the width to use for printing
    """
    print('---'.rjust(width, ' '))
    dots = no_of_dots("total", width)
    print("total" + "."*dots + ":" + str(total).rjust(3))
    print('==='.rjust(width, ' '))
    
def calculate_total(items):
    total = 0
    for v in items.values():
        total = total + v
    return total    
    
    
print_inventory({'pots':5, 'eggs':100}, 'Kitchen', 15)
print('\n')
print_inventory({'pens': 85, 'books': 3, 'pencils': 22}, 'Book Shop', 25)
```

```python
class Animal:

    def __init__(self, name):
        self.name = name

    def info(self):
        print("I'm a", self.name)


class Bird(Animal):

    def move(self):
        print("I'm flying")

    def info(self):
        Animal.info(self)
        print("I have feathers")


class FlightlessBird(Bird):
    
    def __init__(self, info):
        super().__init__(info)
        
    def move(self):
        print("I'm waddling")

crow = Bird('Crow')
crow.info()
crow.move()
penguin = FlightlessBird('Penguin')
penguin.info()
penguin.move()
```

```python
import sys
import csv
from io import StringIO

items = [] # global list

class Error(Exception):
    """Base class for other exceptions"""
    pass

class InvalidInputError(Error):
    """Raised when the input is blank"""
    pass

class ToDo:
    """ToDo item class that creates a basic item entry in the T800"""
    progress = 0 # class-level variable
    TYPE_KEY = 'T'

    def __init__(self, description, status):
        self.description = description
        self.is_done = status
        
    def __str__(self):        
        return self.__status_as_icon().center(6) + ' ' + '| ' + ' '.center(5) \
    + '| ' + self.description.ljust(27) + ' ' + '| ' + '-'

    def mark_as_done(self):
        if not self.is_done: # increment progress if needed
            ToDo.progress = ToDo.progress + 1
        self.is_done = True

    def mark_as_pending(self):
        if self.is_done: # decrement progress if needed
            if ToDo.progress != 0:
                ToDo.progress = ToDo.progress - 1
        self.is_done = False

    def __status_as_icon(self):
        return 'X' if self.is_done else '-'
    
    def as_csv(self):
        """ Return the details of todo object as a list,
        suitable to be stored in a csv file.
        """
        return [self.TYPE_KEY, self.description, 'done' if self.is_done else 'pending']
    
class Deadline (ToDo):
    """Deadline item class that creates a item entry with deadline in the T800"""
    progress = 0 # class-level variable
    TYPE_KEY = 'D'

    def __init__(self, description, status, by):
        super().__init__(description, status)
        self.by = by        
    
    def __str__(self):
        s = super().__str__()
        return s[:-1] + self.by
        
    def as_csv(self):
        c = super().as_csv()
        c.append(self.by)
        return c
        
    def mark_as_done(self):
        if not self.is_done: # increment progress if needed
            Deadline.progress = Deadline.progress + 1
        self.is_done = True
        
    def mark_as_pending(self):
        if self.is_done: # decrement progress if needed
            if Deadline.progress != 0:
                Deadline.progress = Deadline.progress - 1
        self.is_done = False
    
class UserInterface:
    userInput = ""
    
    def __init__(self):
        return
        
    def show_greeting(self):
        print("""
  ______           __      ______                    _             __            
 /_  __/___ ______/ /__   /_  __/__  _________ ___  (_)___  ____ _/ /_____  _____
  / / / __ `/ ___/ //_/    / / / _ \/ ___/ __ `__ \/ / __ \/ __ `/ __/ __ \/ ___/
 / / / /_/ (__  ) ,<      / / /  __/ /  / / / / / / / / / / /_/ / /_/ /_/ / /    
/_/  \__,_/____/_/|_|    /_/  \___/_/  /_/ /_/ /_/_/_/ /_/\__,_/\__/\____/_/     
                                                                                 
""")
        print(">>> Hello, I am the task terminator T-800")
        return
    
    def read_command(self):
        print(">>> What can I do for you?\n")
        return input()
        
    def display(self, screenOutput):
        print(screenOutput)
        
    def print_help(self):
        print(""">>> I'm glad you asked. Here it is:
==================================================
T800 can understand the following commands:

  todo DESCRIPTION 
    Adds a task to the list
    Example: todo read book
  deadline DESCRIPTION "by:" DEADLINE
    Adds a task with deadline to the list
    Example: deadline submit assignment by: end of May
  done INDEX
    Marks the task at INDEX as 'done'
    Example: done 1
  delete INDEX
    Deletes item at the index
    Example: delete 1
  pending INDEX
    Reverts done item at index to 'pending'
    Example: pending 1
  exit
    Exits the application
  help
    Shows the help information
  list
    Lists the tasks in the list
  progress
    Shows the progress of the current tasks
--------------------------------------------------""")
    
class StorageManager:
    def __init__(self, filename):
        self.filename = filename

class TaskManager:    
    def __init__(self, fileObject):
        self.fileObject = fileObject
        self.filename = fileObject.filename
        
    def __create_file_if_missing(self):
        open(self.fileObject.filename, 'a').close()
    
    def load_data(self):
        TaskManager.__create_file_if_missing(self)
        with open(self.fileObject.filename, 'r') as csvfile:
            file_handler = csv.reader(csvfile)
            for row in file_handler:
                if not row:
                    continue
                TaskManager.__load_item_from_csv_line(row)
            return
        
    def __load_item_from_csv_line(row):
        if row[0] == 'T':
            items.append(ToDo(row[1], True if row[2] == 'True' else False))
        elif row[0] == 'D':
            items.append(Deadline(row[1], True if row[2] == 'True' else False, row[3]))
        return
    
    def save_data(self):
        with open(self.fileObject.filename, "w", newline='') as csvfile:
            output = csv.writer(csvfile)
            for item in items:
                if isinstance(item, Deadline):          
                    output_to_file = ["D",item.description,item.is_done,item.by]
                else:
                    output_to_file = ["T",item.description,item.is_done]
                output.writerow(output_to_file)
                
    def get_items_as_table(self):
        normal_stdout = sys.stdout # store old sysout
        result = StringIO()
        sys.stdout = result
        
        TaskManager.print_items()
        
        sys.stdout = normal_stdout # restore old sysout
        return result.getvalue()
    
    def print_items():
        if len(items) == 0:
            print('>>> Error 404: Nothing to list')
        else:
            print(">>> Here is the list of tasks:")
            print("""============================================================
STATUS | INDEX | DESCRIPTION                 | DEADLINE
------------------------------------------------------------""")
            TaskManager.__print_elements()
            print("------------------------------------------------------------")
        return
    
    def __print_elements():
        for i, element in enumerate(items):
            print(str(element)[:11] + str(i+1) + str(element)[11:])
        return
        
    def add_item(self, user_input):
        command_parts = user_input.strip().split(' ', 1)
        try:
            items.append(ToDo(command_parts[1], False))
            return ("New item: " + "'" + command_parts[1] + "'" + " added")
        except IndexError:
                raise IndexError("INPUT: todo \"task\"")
        
    def add_deadline_item(self, user_input):
        command_parts = user_input.strip().split(' ', 1)
        try:
            due = command_parts[1].partition("by:")[2].strip()
            task = command_parts[1].partition("by:")[0].strip()
            if due == "" or task == "":
                raise InvalidInputError
            items.append(Deadline(task, False, due))
            return ("New item: " + "'" + task + "'" + " added. " + "Deadline: " + "'" + due + "'")
        except InvalidInputError:
            raise InvalidInputError("INPUT: deadline \"task\" by: \"due date\"")
        except IndexError:
            raise IndexError("No deadline task provided")

    def mark_item_as_done(self, user_input):
        index_as_string = user_input[5:].strip()
        index_to_remove = TaskManager.__index_check(index_as_string)
        for i, obj in enumerate(items):
            if i == index_to_remove:
                if obj.is_done:
                    return ("Item: " + "'" + obj.description + "'" + " has been done already!")
                obj.mark_as_done()
                return ("Item: " + "'" + obj.description + "'" + " marked as done")

    def mark_item_as_pending(self, user_input):
        index_as_string = user_input[8:].strip()
        index_to_remove = TaskManager.__index_check(index_as_string)
        for i, obj in enumerate(items):
            if i == index_to_remove:
                if not obj.is_done:
                    return ("Item: " + "'" + obj.description + "'" + " is already pending!")
                obj.mark_as_pending()
                return ("Item: " + "'" + obj.description + "'" + " marked as pending")
    
    def __index_check(string):
        try:
            index = int(string.strip())
        except Exception:
            raise ValueError(string + ' is not a number')
        if index < 1:
            raise ValueError('Index must be greater than 0')
        try:
            if items[index - 1]:
                return index - 1
        except IndexError:
            raise ValueError('No item at index ' + string)
    
    def delete_item(self, user_input):
        try:
            s = int(user_input[7:].strip())
            if s == 0:
                raise IndexError
            deleted_item = items.pop(s-1)
            return ("Task: " + "'" + deleted_item.description + "'" + " deleted from the list")
        except IndexError:
            raise IndexError("There is no list item at the number you typed!")
        except ValueError:
            raise ValueError("Only integers accepted as input")
            
    def get_current_progress(self):
        status = {'Todo': 0, 'Deadline': 0}
        tFlag = 0
        dFlag = 0

        for obj in items:
            if isinstance(obj, Deadline) and dFlag == 0:
                status['Deadline'] = obj.progress
                dFlag = 1
            elif tFlag == 0:
                status['Todo'] = obj.progress
                tFlag = 1
        return(">>> Progress for this session: todos {} deadlines {}".format(status['Todo'], status['Deadline']))
    
    def mass(self, user_input):
        command = user_input.strip()
        if command[7:].startswith('delete'):
            return ("apple")
#             print(command[7:])
                
def execute_command(command_parts, task_manager, ui):
    command = command_parts.strip()
    if command == '':
        return
    elif command == 'exit':
        ui.display('Bye!')
        sys.exit()
    elif command == 'list':
        message = task_manager.get_items_as_table()
        ui.display(message)
    elif command == 'help':
        ui.print_help()
    elif command == 'progress':
        message = task_manager.get_current_progress()
        ui.display(message)
        
    elif command.startswith('todo'):
        message = task_manager.add_item(command)
        ui.display(message)
    elif command.startswith('deadline'):
        message = task_manager.add_deadline_item(command)
        ui.display(message)
    elif command.startswith('done'):
        message = task_manager.mark_item_as_done(command)
        ui.display(message)
    elif command.startswith('pending'):
        message =  task_manager.mark_item_as_pending(command)
        ui.display(message)
    elif command.startswith('delete'):
        message = task_manager.delete_item(command)
        ui.display(message)
    elif command.startswith('mass'):
        message = task_manager.mass(command)
        ui.display(message)
    else:
        raise Exception('Command not recognized')
    return


def main():
    # create a UserInterface object and show the greeting
    ui = UserInterface()
    ui.show_greeting()

    # create a StorageManager object
    storage = StorageManager('monty7.csv')

    # create a TaskManager object and load data
    task_manager = TaskManager(storage)
    task_manager.load_data()

    while True:
        try:
            command = ui.read_command()
            execute_command(command, task_manager, ui)
            task_manager.save_data()
        except Exception as e:
            ui.display('SORRY, I could not perform that command. Problem: ' + str(e))

main()
```

```python
def string_splitter(arr, string, split_length):
    if len(string) < split_length:
        arr.append(string)
        return
    else:
        arr.append(string[:split_length])
        return string_splitter(arr, string[split_length:], split_length)
```
