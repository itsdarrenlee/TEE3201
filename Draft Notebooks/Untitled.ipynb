```python
import sys
import csv
import copy

filename = 'monty7.csv'
items = []

class ToDo:
    def __init__(self, description, status):
        self.description = description
        self.is_done = status

    def mark_as_done(self):
        self.is_done = True

    def mark_as_pending(self):
        self.is_done = False

    def as_string(self):
        """ Return the details of todo object as a string"""
        status = 'X' if self.is_done else '-'
        return status.center(6) + ' ' + '| ' + ' '.center(5) \
    + '| ' + self.description.ljust(27) + ' ' + '| ' + '-'

    def as_csv(self):
        """ Return the details of todo object as a list,
        suitable to be stored in a csv file.
        """
        return ['T', self.description, 'done' if self.is_done else 'pending']
    
class Deadline(ToDo):
    def __init__(self, description, status, deadline):
        self.deadline = deadline
        ToDo.__init__(self, description, status)
    
    def as_csv(self):
        """ Return the details of deadline object as a list,
        suitable to be stored in a csv file.
        """
        return ['D', self.description, 'done' if self.is_done else 'pending', self.deadline]
    
    def as_string(self):
        """ Return the details of deadline object as a string"""
        status = 'X' if self.is_done else '-'
        return status.center(6) + ' ' + '| '  + ' '.center(5) \
    +  '| ' + self.description.ljust(27) + ' ' + '| ' + self.deadline


def print_help():
    print(""">>> I'm glad you asked. Here it is:
==================================================
T800 can understand the following commands:

  todo DESCRIPTION 
    Adds a task to the list
    Example: todo read book
  deadline DESCRIPTION "by:" DEADLINE
    Adds a task with deadline to the list
    Example: deadline submit assignment by: end of May
  done INDEX
    Marks the task at INDEX as 'done'
    Example: done 1
  delete INDEX
    Deletes item at the index
    Example: delete 1
  pending INDEX
    Reverts done item at index to 'pending'
    Example: pending 1
  exit
    Exits the application
  help
    Shows the help information
  list
    Lists the tasks in the list
--------------------------------------------------""")
    return

def add_item_from_csv_line(row):
    if row[0] == 'T':
        items.append(ToDo(row[1], True if row[2] == 'TRUE' else False))
    elif row[0] == 'D':
        items.append(Deadline(row[1], True if row[2] == 'TRUE' else False, row[3]))
    return

def load_data(filename):
    with open(filename, 'r') as csvfile:
        deliveries_reader = csv.reader(csvfile)
        for row in deliveries_reader:
            if not row:
                continue
            add_item_from_csv_line(row)
        return

def print_elements():
    itemsCopy = copy.deepcopy(items)
    for i, element in enumerate(itemsCopy):
        s = element.as_string()
        print(s[:11] + str(i+1) + s[11:])
    return

def print_items():
    if len(items) == 0:
        print('>>> Error 404: Nothing to list')
    else:
        print(">>> Here is the list of tasks:")
        print("""============================================================
STATUS | INDEX | DESCRIPTION                 | DEADLINE
------------------------------------------------------------""")
        print_elements()
        print("------------------------------------------------------------")
    return
    
def print_help():
    print(""">>> I'm glad you asked. Here it is:
==================================================
T800 can understand the following commands:

  todo DESCRIPTION 
    Adds a task to the list
    Example: todo read book
  deadline DESCRIPTION "by:" DEADLINE
    Adds a task with deadline to the list
    Example: deadline submit assignment by: end of May
  done INDEX
    Marks the task at INDEX as 'done'
    Example: done 1
  delete INDEX
    Deletes item at the index
    Example: delete 1
  pending INDEX
    Reverts done item at index to 'pending'
    Example: pending 1
  exit
    Exits the application
  help
    Shows the help information
  list
    Lists the tasks in the list
--------------------------------------------------""")
    return

def print_greeting():
    print("""
 _   _        _          __         _                          
| | | |      | |        / _|       | |                         
| |_| |  ___ | | _ __  | |_  _   _ | |                         
|  _  | / _ \| || '_ \ |  _|| | | || |                         
| | | ||  __/| || |_) || |  | |_| || |                         
\_| |_/ \___||_|| .__/ |_|   \__,_||_|                         
                | |                                            
                |_|                                            
 _                           _                _                
| |                         (_)              | |               
| |_   ___  _ __  _ __ ___   _  _ __    __ _ | |_   ___   _ __ 
| __| / _ \| '__|| '_ ` _ \ | || '_ \  / _` || __| / _ \ | '__|
| |_ |  __/| |   | | | | | || || | | || (_| || |_ | (_) || |   
 \__| \___||_|   |_| |_| |_||_||_| |_| \__,_| \__| \___/ |_|   
                                                               
                                                               

""")
    print(">>> Hello, I am the terminator T-800")
    return

def create_file_if_missing(filename):
    open(filename, 'a').close()
    
def save_data(filename, items):
    with open(filename, "w", newline='') as csvfile:
        output = csv.writer(csvfile)
        for item in items:
            if isinstance(item, Deadline):          
                output_to_file = ["D",item.description,item.is_done,item.deadline]
            else:
                output_to_file = ["T",item.description,item.is_done]
            output.writerow(output_to_file)

def add_item(user_input):
    command_parts = user_input.strip().split(' ', 1)
    try:
        items.append(ToDo(command_parts[1], False))
    except AttributeError as ATBERR:
        print(ATBERR)
    return

def add_deadline_item(user_input):
    command_parts = user_input.strip().split(' ', 1)
    try:
        due = command_parts[1].partition("by:")[2].strip()
        task = command_parts[1].partition("by:")[0].strip()
        items.append(Deadline(task, False, due))
    except AttributeError as ATBERR:
        print(ATBERR)    
    return

def delete_item(user_input):
    try:
        s = int(user_input[7:].strip())
        items.pop(s-1)
        print("Task deleted from the list")
    except IndexError as IDX:
        print("Character you typed is out of range! Error: " + str(IDX))
    except ValueError as VALERR:
        print("You typed an invalid character! Error: " + str(VALERR))
    return  

def index_check(string):
    try:
        index = int(string.strip())
    except Exception:
        raise ValueError(string + ' is not a number')
    if index < 1:
        raise ValueError('Index must be greater than 0')
    try:
        if items[index - 1]:
            return index - 1
    except IndexError:
        raise ValueError('No item at index ' + string)
    return

def mark_item_as_done(user_input):
    index_as_string = user_input[5:].strip()
    index_to_remove = index_check(index_as_string)
    for i, obj in enumerate(items):
        if i == index_to_remove:
            obj.mark_as_done()
    return
  
def mark_item_as_pending(user_input):
    index_as_string = user_input[8:].strip()
    index_to_remove = index_check(index_as_string)
    for i, obj in enumerate(items):
        if i == index_to_remove:
            obj.mark_as_done()
    return

def read_command():
    print(">>> What can I do for you?")
    print()
    return input()

def terminate():
    print(">>> Hasta La Vista, Baby!")
    sys.exit()

def execute_command(command):
    command = command.strip()
    if command == '':
        return
    elif command == 'exit':
        terminate()
    elif command == 'list':
        print_items()
    elif command == 'help':
        print_help()
    elif command.startswith('todo '):
        add_item(command)
    elif command.startswith('deadline '):
        add_deadline_item(command)
    elif command.startswith('done '):
        mark_item_as_done(command)
    elif command.startswith('pending '):
        mark_item_as_pending(command)
    elif command.startswith('delete '):
        delete_item(command)
    else:
        raise Exception('Command not recognized')
    return

def main():
    create_file_if_missing(filename)
    load_data(filename) # load task data from the file
    print_greeting()
    while True:
        try:
            command = read_command()
            execute_command(command)
            save_data(filename, items) # save all tasks in the file
        except Exception as e:
            print('>>> SORRY, I could not perform that command. Problem:', e)

main()
```
